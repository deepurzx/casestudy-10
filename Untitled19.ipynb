```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
data['class'] = le.fit_transform(data['class'])
data
```

```python
#spiting dataset in to feature and target
x=data.drop('class',axis=1)
y=data['class']
```

```python
x = pd.get_dummies(x)
```
